```python
from skimage import color, io

# Load an RGB image using scikit-image's `io.imread()` function
rgb_image = io.imread('data/test_black/image5005.jpg')

# Convert the RGB image to the CIELAB color space using `color.rgb2lab()`
lab_image = color.rgb2lab(rgb_image)

# Display the original and converted images side by side
io.imshow_collection([rgb_image, lab_image])
io.show()
```
